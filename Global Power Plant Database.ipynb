```python
countries_plant = df.country_long.value_counts().head(20)
countries_plant
```

```python
import matplotlib
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12, 8)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
sns.barplot(x = countries_plant.index, y = countries_plant)
plt.xticks(rotation = 90)
plt.title('Country Designation')
plt.ylabel('Number of Power Plant')
plt.xlabel('Countries');
```

```python
main_primary_fuel = df.primary_fuel.value_counts() * 100 / df.primary_fuel.count()
main_primary_fuel
```

Various types of fuel are used in the power plant, which can be classified as primary fuel, other fuel-1, other fuel-2, etc. Primary fuel is used to run the main turbine to generate electricity and other fuels are used to run auxiliary machines to support the generation process.

```python
sns.barplot(x = main_primary_fuel, y = main_primary_fuel.index)
plt.title('Main primary fuel')
plt.xlabel('Count (Percentages)');
plt.ylabel('Different type of power plant depands on the type of fuel uses');
```

Renewable energy-based power plants have higher percentages that are beneficial to our future generation. Global warming is the next major threat that we all have to face. It is a good sign that the use of coal, hydro & solar is increasing.

```python
sns.scatterplot(x = df.capacity_mw, y =df.primary_fuel, s = 150)
plt.title('Type of power plant and capacity');
```

The Coal Power Plant has most of the high-capacity power plant. There is only 0.67 per cent of the Biomass power plant, but the capacity of the power plant is high. In spite of having a higher percentage, the solar power plant has a small capacity.
